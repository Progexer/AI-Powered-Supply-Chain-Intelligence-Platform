# SupplyChainIQ Exploratory Data Analysis

This notebook profiles the cleaned datasets, creates visualizations, and captures business-relevant findings for demand, delivery, supplier, inventory, and cost patterns.

```python
from pathlib import Path
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style='whitegrid')
base = Path('data/processed')
charts = Path('notebooks/charts_eda')
charts.mkdir(parents=True, exist_ok=True)

d1 = pd.read_csv(base / 'supplychainiq_dataset1_clean.csv')
d2 = pd.read_csv(base / 'supplychainiq_dataset2_clean.csv')

for col in ['order_date', 'shipping_date']:
    if col in d1.columns:
        d1[col] = pd.to_datetime(d1[col], errors='coerce')

d1.head()
```

```python
# Core charts used in the business report.
fig, ax = plt.subplots(figsize=(10, 6))
d1['shipping_mode'].value_counts().sort_values(ascending=False).plot(kind='bar', ax=ax, color='#2563eb')
ax.set_title('Dataset 1: Shipping Mode Distribution')
ax.set_xlabel('Shipping Mode')
ax.set_ylabel('Orders')
fig.tight_layout()
fig.savefig(charts / 'd1_shipping_mode_distribution.png', dpi=160)
plt.close(fig)

fig, ax = plt.subplots(figsize=(10, 6))
d1['order_region'].value_counts().head(10).sort_values().plot(kind='barh', ax=ax, color='#0f766e')
ax.set_title('Dataset 1: Top Order Regions')
ax.set_xlabel('Orders')
ax.set_ylabel('Region')
fig.tight_layout()
fig.savefig(charts / 'd1_top_regions.png', dpi=160)
plt.close(fig)

fig, ax = plt.subplots(figsize=(10, 6))
d1['category_name'].value_counts().head(10).sort_values().plot(kind='barh', ax=ax, color='#7c3aed')
ax.set_title('Dataset 1: Top Product Categories')
ax.set_xlabel('Line Items')
ax.set_ylabel('Category')
fig.tight_layout()
fig.savefig(charts / 'd1_top_categories.png', dpi=160)
plt.close(fig)

fig, ax = plt.subplots(figsize=(10, 6))
d2['product_type'].value_counts().sort_values(ascending=False).plot(kind='bar', ax=ax, color='#db2777')
ax.set_title('Dataset 2: Product Type Distribution')
ax.set_xlabel('Product Type')
ax.set_ylabel('SKUs')
fig.tight_layout()
fig.savefig(charts / 'd2_product_type_distribution.png', dpi=160)
plt.close(fig)

fig, ax = plt.subplots(figsize=(10, 6))
d2['inspection_result'].value_counts().reindex(['Pass', 'Fail', 'Pending']).plot(kind='bar', ax=ax, color='#ea580c')
ax.set_title('Dataset 2: Inspection Results')
ax.set_xlabel('Inspection Result')
ax.set_ylabel('SKUs')
fig.tight_layout()
fig.savefig(charts / 'd2_inspection_results.png', dpi=160)
plt.close(fig)

fig, ax = plt.subplots(figsize=(10, 6))
sns.histplot(d2['defect_rate_raw'], bins=15, kde=True, ax=ax, color='#dc2626')
ax.set_title('Dataset 2: Defect Rate Distribution')
ax.set_xlabel('Defect Rate (raw)')
ax.set_ylabel('SKU Count')
fig.tight_layout()
fig.savefig(charts / 'd2_defect_rate_distribution.png', dpi=160)
plt.close(fig)
```

## Business Intelligence Use Cases

- Identify demand concentration by region and category.
- Measure shipment reliability by shipping mode.
- Spot supply-side quality risk from inspection and defect-rate patterns.
- Feed feature selection for forecasting, delay prediction, and supplier scoring.
